```python
fruits_input = input("Enter three of your favorite fruits (comma-separated): ")
```

```python
fruit_list = [fruit.strip() for fruit in fruits_input.split(",")]

print("Your list of fruits:", fruit_list)
```

```python
print("First fruit:", fruit_list[0])
print("Last fruit:", fruit_list[-1])
print("Second fruit:", fruit_list[1])
print("First two fruits:", fruit_list[:2])
```

```python
new_fruit = input("Enter a new fruit to add: ")
fruit_list.append(new_fruit.strip())
print("Updated list after adding:", fruit_list)
```

```python
remove_fruit = input("Enter a fruit to remove: ")
if remove_fruit in fruit_list:
    fruit_list.remove(remove_fruit)
    print("Updated list after removing:", fruit_list)
else:
    print("Fruit not found in list.")

print("Length of the list is now:", len(fruit_list))
```

```python
letters_list = list(full_name.replace(" ", ""))

print("Letters list:", letters_list)
print("First 3 letters:", letters_list[:3])
print("Last 3 letters:", letters_list[-3:])
```

```python
modified_letters = []

for i in range(len(letters_list)):
    if i < 3:
        modified_letters.append(letters_list[i].upper())
    else:
        modified_letters.append(letters_list[i].lower())

modified_name = "".join(modified_letters)

print("Modified name:", modified_name)
```

```python
numbers_input = input("Enter 5 numbers (space-separated): ")

numbers_list = [int(num) for num in numbers_input.split()]

total = sum(numbers_list)
average = total / len(numbers_list)

print("Sum of your numbers:", total)
print("Average of your numbers:", average)
```
